```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/ENJOYSPORT.csv')
data
```

```python
hypothesis = ['@'] * (len(data.columns) - 1)
hypothesis
```

```python
for i in range(len(data)):
  x = list(data.iloc[i])
  print(x)
```

```python
def FindS(H,x):
  if((x[len(x)-1]) == 1):
    for i in range(len(x)-1):
      if H[i] == '@':
        H[i] = x[i]
      elif H[i] != x[i]:
        H[i] = '?'
```

```python
for i in range(len(data)):
  x = list(data.iloc[i])
  FindS(hypothesis,x)
```
