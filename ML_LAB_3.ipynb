```python
## Height
x = [5.0, 5.1, 5.2, 5.3, 5.4, 5.5, 5.6, 5.7, 5.8]
## Weight
y = [53, 55, 59, 61, 65, 68, 70, 74, 76]
```

```python
meanx = sum(x) / len(x)
meany = sum(y) / len(y)
```

```python
nm = sum((xi - meanx) * (yi - meany) for xi,yi in zip(x,y))
dm = sum((xi - meanx) ** 2 for xi in x)
```

```python
w1 = nm / dm
```

```python
w0 = meany - w1*meanx
```

```python
print(f"y = {w1:.2f}x + {w0:.2f}")
```

```python
def value1(x):
  return w1*x + w0

inpx = 5.9
predicty = value1(inpx)
print(f"Predicted Value of Y for x = {inpx}: {predicty:.2f}")
```

```python
ploty = [value1(xi) for xi in x]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='green', label='Original data')
plt.plot(x, ploty, color='blue', label='Fitted line')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()
```
